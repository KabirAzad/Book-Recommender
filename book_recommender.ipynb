```python
import numpy as np
import pandas as pd
import math
```

```python
books = pd.read_csv('/content/Books.csv')
users = pd.read_csv('/content/Users.csv')
ratings = pd.read_csv('/content/Ratings.csv')
```

```python
rating_with_name = ratings.merge(books,on='ISBN')
```

```python
rating_with_name
```

```python
count_rating = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
count_rating.rename(columns={'Book-Rating':'rating_count'},inplace=True)
count_rating
```

```python
avg_rating = rating_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns={'Book-Rating':'Average_Rating'},inplace=True)
```

```python
avg_rating.head()
```

```python
final_df = count_rating.merge(avg_rating,on='Book-Title')
final_df.head()
```

```python
final_df = final_df[final_df['rating_count']>=250].sort_values('Average_Rating',ascending=False)
```
